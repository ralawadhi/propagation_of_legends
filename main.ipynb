```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create population
# 1:Yay, 2:Nay, 3:Neutral, 0.45,0.45,0.10
probability_distribution = [0.15,0.15,0.70]
arch_type = [1,2,3]
pop_size = 100000
pop_list = []
#populate
for i in range(pop_size):
    selected_type = np.random.choice(arch_type, p = probability_distribution)
    pop_list.append(selected_type)
```

```python
plt.hist(pop_list)
```

```python
initial_thing = 3
thing_evolution = [initial_thing]
for i in pop_list:
    if initial_thing > 0:
        if i == 1:
            if np.random.random() <= 0.05:
                initial_thing = initial_thing + 1
                thing_evolution.append(initial_thing)
            else:
                thing_evolution.append(initial_thing)
        elif i == 2:
            if np.random.random() <= 0.05:
                initial_thing = initial_thing - 1
                thing_evolution.append(initial_thing)
            else:
                thing_evolution.append(initial_thing)
        else:
            thing_evolution.append(initial_thing)
    else:
        initial_thing =+ 1
```

```python
gen_list = []
thing_list = []
for gen, thing in list(enumerate(thing_evolution)):
    gen_list.append(gen)
    thing_list.append(thing)
```

```python
#moving average

def moving_avg(array, window_size):
    moving_averages = []
    i = 0
    while i < len(array) - window_size + 1:
        window = array[i:i+window_size]
        window_average = round(sum(window)/window_size)
        moving_averages.append(window_average)
        i+=1
    
    return moving_averages

moving_average = moving_avg(thing_list, 10000)
```

```python
plt.plot(gen_list, thing_list)
plt.plot(moving_average, linewidth=3)
plt.xlabel("Generation")
plt.ylabel("I saw N things!")
```
